```python
#https://medium.com/analytics-vidhya/image-classification-using-machine-learning-support-vector-machine-svm-dc7a0ec92e01

### load training data

import pandas as pd
import os
from skimage.transform import resize
from skimage.io import imread
import numpy as np
import matplotlib.pyplot as plt
import random

domain=['target','source']
flat_data_arr=[] #input array
target_arr=[] #output array
datadir='../../data/denoising/train' 

for i in domain:
    print(f'loading... domain : {i}')
    if i=='target':
        path=os.path.join(datadir,'mayo/full_1mm')
        for root, dirs, filenames in os.walk(path):
            for filename in filenames:
                print(filename)
                img_array = imread(os.path.join(root,filename))
                sh = random.randint(0,431)
                sw = random.randint(0,431)
                img_crop = img_array[sh:sh+80,sw:sw+80]
                flat_data_arr.append(img_crop.flatten())
                target_arr.append(1) # set target domain as 1
                
        path=os.path.join(datadir,'mayo/full_3mm')
        for root, dirs, filenames in os.walk(path):
            for filename in filenames:
                print(filename)
                img_array = imread(os.path.join(root,filename))
                sh = random.randint(0,431)
                sw = random.randint(0,431)
                img_crop = img_array[sh:sh+80,sw:sw+80]
                flat_data_arr.append(img_crop.flatten())
                target_arr.append(1) # set target domain as 1
                
        path=os.path.join(datadir,'mayo/quarter_1mm')
        for root, dirs, filenames in os.walk(path):
            for filename in filenames:
                print(filename)
                img_array = imread(os.path.join(root,filename))
                sh = random.randint(0,431)
                sw = random.randint(0,431)
                img_crop = img_array[sh:sh+80,sw:sw+80]
                flat_data_arr.append(img_crop.flatten())
                target_arr.append(1) # set target domain as 1
                
        path=os.path.join(datadir,'mayo/quarter_3mm')
        for root, dirs, filenames in os.walk(path):
            for filename in filenames:
                print(filename)
                img_array = imread(os.path.join(root,filename))
                sh = random.randint(0,431)
                sw = random.randint(0,431)
                img_crop = img_array[sh:sh+80,sw:sw+80]
                flat_data_arr.append(img_crop.flatten())
                target_arr.append(1) # set target domain as 1

    else:
        path=os.path.join(datadir,'phantom/ge/chest/level3_025_crop320')
        for img in os.listdir(path):
            print(img)
            img_array = imread(os.path.join(path,img))
            sh = random.randint(0,239)
            sw = random.randint(0,239)
            img_crop = img_array[sh:sh+80,sw:sw+80]
            flat_data_arr.append(img_crop.flatten())
            target_arr.append(0) # set source domain as 0
            
        path=os.path.join(datadir,'phantom/ge/pelvis/level3_025_crop320')
        for img in os.listdir(path):
            print(img)
            img_array = imread(os.path.join(path,img))
            sh = random.randint(0,239)
            sw = random.randint(0,239)
            img_crop = img_array[sh:sh+80,sw:sw+80]
            flat_data_arr.append(img_crop.flatten())
            target_arr.append(0) # set source domain as 0
            
        path=os.path.join(datadir,'phantom/ge/chest/level5_005_crop320')
        for img in os.listdir(path):
            print(img)
            img_array = imread(os.path.join(path,img))
            sh = random.randint(0,239)
            sw = random.randint(0,239)
            img_crop = img_array[sh:sh+80,sw:sw+80]
            flat_data_arr.append(img_crop.flatten())
            target_arr.append(0) # set source domain as 0
            
        path=os.path.join(datadir,'phantom/ge/pelvis/level5_005_crop320')
        for img in os.listdir(path):
            print(img)
            img_array = imread(os.path.join(path,img))
            sh = random.randint(0,239)
            sw = random.randint(0,239)
            img_crop = img_array[sh:sh+80,sw:sw+80]
            flat_data_arr.append(img_crop.flatten())
            target_arr.append(0) # set source domain as 0


    print(f'loaded category:{i} successfully')

flat_data=np.array(flat_data_arr)
target=np.array(target_arr)
df=pd.DataFrame(flat_data) #dataframe
df['domain']=target
x_train=df.iloc[:,:-1] #input data 
y_train=df.iloc[:,-1] #output data
print(df)
```

```python
from sklearn import svm

svc = svm.SVC(C=10, kernel='rbf', gamma=0.001, probability=True)
```

```python
svc.fit(x_train,y_train).score(x_train,y_train)
print('The Model is trained well with the given images')
```

```python
# train accuracy

from sklearn.metrics import accuracy_score

probability=svc.predict_proba(x_train)
print(probability)

prob = []
for i in range(probability.shape[0]):
    prob.append(probability[i,np.array(y_train)[i]])
prob = np.array(prob)
print("The estimated domain is:")
print(prob)

y_pred=svc.predict(x_train)
print("The predicted data is:")
print(y_pred)
print("The actual data is:")
print(np.array(y_train))
print(f"The model is {accuracy_score(y_pred,y_train)}% accurate")
```

```python
# train loss

import math

loss=0

for i in range(prob.shape[0]):
    loss -= math.log10(prob[i])

loss = loss/prob.shape[0]
print(loss)
```

```python
# saving model

import pickle
import joblib

saved_model = joblib.dump(svc,'./tests/svc2.pkl')
```

```python
### load test data (source)

import pandas as pd
import os
from skimage.transform import resize
from skimage.io import imread
import numpy as np
import matplotlib.pyplot as plt
import random

domain='source'
flat_data_arr=[] #input array
target_arr=[] #output array
datadir='../../data/denoising/test_result_DA'

path=os.path.join(datadir,"ge-20210721-0132-rev-edsr-p-chest-pelvis-testset-ge-['chest', 'pelvis']")

print(f'loading... domain : source')

for root, dirs, filenames in os.walk(path):
    for filename in filenames:
        print(filename)
        img_array = imread(os.path.join(root,filename))
        sh = random.randint(0,239)
        sw = random.randint(0,239)
        img_crop = img_array[sh:sh+80,sw:sw+80]
        flat_data_arr.append(img_crop.flatten())
        target_arr.append(0) # set source domain as 0
    break

flat_data=np.array(flat_data_arr)
target=np.array(target_arr)
df=pd.DataFrame(flat_data) #dataframe
df['domain']=target
x_test=df.iloc[:,:-1] #input data 
y_test=df.iloc[:,-1] #output data
print(df)
```

```python
# load model for test

import joblib

loaded_svc = joblib.load('./tests/svc2.pkl')
```

```python
# test accuracy

from sklearn.metrics import accuracy_score

probability=loaded_svc.predict_proba(x_test)
print(probability)

prob = []
for i in range(probability.shape[0]):
    prob.append(probability[i,np.array(y_test)[i]])
prob = np.array(prob)
print("The estimated domain is:")
print(prob)

y_pred=loaded_svc.predict(x_test)
print("The predicted data is:")
print(y_pred)
print("The actual data is:")
print(np.array(y_test))
print(f"The model is {accuracy_score(y_pred,y_test)} accurate")
```
